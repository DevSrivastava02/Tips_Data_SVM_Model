```python
#  Independent and dependent features
X=df.drop("total_bill",axis=1)
y=df["total_bill"]
```

```python
#  Train to split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
X_train.head()
```

####  feature encoding
#### Label encoding and one hot encoding
####  if 2 category we use label
####  more than 2 we use one hot encoding

```python
from sklearn.preprocessing import LabelEncoder
leb1=LabelEncoder()
leb2=LabelEncoder()
leb3=LabelEncoder()
```

```python
X_train["sex"]=leb1.fit_transform(X_train["sex"])
X_train["smoker"]=leb2.fit_transform(X_train["smoker"])
X_train["time"]=leb3.fit_transform(X_train["time"])

#  Test dataset encoding
X_test["sex"]=leb1.fit_transform(X_test["sex"])
X_test["smoker"]=leb2.fit_transform(X_test["smoker"])
X_test["time"]=leb3.fit_transform(X_test["time"])
```

One hot encoding and columns tranformation

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])],remainder="passthrough")
X_train=ct.fit_transform(X_train)
X_test=ct.transform(X_test)
```

```python
#  Standard sclear
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
X_train=scalar.fit_transform(X_train)
X_test=scalar.transform(X_test)
```

```python
#  MOdel training
from sklearn.svm import SVR
svr=SVR()
```

```python
#  fit x_train and y_train
svr.fit(X_train,y_train)
```

```python
#  prediction the data
y_pred=svr.predict(X_test)
# performance matrix
from sklearn.metrics import mean_absolute_error,r2_score
mse=mean_absolute_error(y_test,y_pred)
print(mse)
score=r2_score(y_test,y_pred)
print(score)
```

Hyperparamter tuning using GridsearchCV

```python
from sklearn.model_selection import GridSearchCV

#  defining parameter range
c_value=[0.1,1,10,100,1000]
gamma=[1,0.1,0.01,0.001,0.0001]
kernal=['linear', 'poly', 'rbf', 'sigmoid']
param={"C":c_value,
       "gamma":gamma,
       "kernel":kernal}

grid=GridSearchCV(estimator=svr,param_grid=param,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
grid.best_params_
grid.best_estimator_
grid.best_params_
grid.best_score_
#  prediction
y_pred=grid.predict(X_test)



```
